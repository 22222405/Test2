```python
import pandas as pd
```

```python
df=pd.read_csv('Settlements.csv')
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

```

```python
avg_size_column = df['avg_hhsize'].tolist()
selection_sort(avg_size_column)
print("Sorted 'avg_hhsize' column:", avg_hhsize_column)
```
