```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Settlements.csv')
```

```python
print(df.info())
```

```python
summary_stats = df.describe()
print(summary_stats)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='pocket_name', y='density_dwellings', data=df, order=df.sort_values('density_dwellings', ascending=False)['pocket_name'])
plt.title('Population Density vs Informal Settlements')
plt.xlabel('Pocket Code')
plt.ylabel('Density of Dwellings')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['avg_hhsize'], bins=20, kde=True)
plt.title('Distribution of Average Household Size')
plt.xlabel('Average Household Size')
plt.ylabel('Frequency')
plt.show()
```

```python
sanitation_counts = df[['toilets_full_flush', 'toilets_chemical', 'toilets_portable', 'toilets_container', 'toilets_other']].sum()
plt.figure(figsize=(8, 8))
plt.pie(sanitation_counts, labels=sanitation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Access to Sanitation Facilities')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='density_dwellings', y='avg_hhsize', data=df)
plt.title('Correlation between Density of Dwellings VS Average Household Size')
plt.xlabel('Density of Dwellings')
plt.ylabel('Average Household Size')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['structure_count'])
plt.title('Distribution of Structure Count')
plt.xlabel('Structure Count')
plt.show()
```
